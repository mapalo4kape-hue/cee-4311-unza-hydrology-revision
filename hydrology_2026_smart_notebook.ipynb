## Meeting 1 Summary — Precipitation, Rainfall Instruments, Missing Data, and Consistency

### Overview
This revision session focused on precipitation concepts, rainfall measurement instruments, methods for estimating missing rainfall data, and rainfall consistency checking.

### Precipitation fundamentals
- Precipitation: water falling from atmosphere to earth when gravity overcomes atmospheric resistance.
- Forms covered: rain, drizzle, snow, hail.
- Size highlights:
  - Drizzle drops: < 0.5 mm
  - Hailstones: ~0.5 to 5 cm
- Practical relevance: irrigation, dams, flood prediction, drainage design, planning, and safety.

### Types of precipitation
- Convective precipitation
- Orographic precipitation
- Cyclonic precipitation

### Rainfall measurement instruments
- Standard non-recording gauge: ordinary rain gauge.
- Recording gauges discussed:
  - Tipping bucket gauge
  - Float-type gauge
  - Weighing-type gauge
  - Self-recording approaches
  - Laser-based gauge (noted)

- Allowable error frequently taken as 10% unless specified.

### Missing rainfall methods
1. Normal ratio method
2. Arithmetic mean method
3. Inverse distance method

### Consistency checking
- Double mass curve method used to detect record inconsistency and apply slope-based correction.

### Mean areal precipitation
- Thiessen polygon method emphasized.
- Isohyetal method also discussed.

### Exam tip
Read wording carefully to decide whether to define, classify, estimate, check consistency, or compute mean areal rainfall.


## Meeting 2 Summary — Infiltration Theory, Factors, and Horton Equation

### Overview
Focused on infiltration definitions, factors controlling infiltration, measurement methods, and solving exam-style Horton equation questions.

### Key definitions
- Infiltration
- Percolation
- Infiltration capacity (maximum under given condition)
- Infiltration rate
- Cumulative infiltration
- Steady-state infiltration

### Why infiltration matters
- Flood prediction and mapping
- Groundwater recharge
- Irrigation and agriculture
- Erosion control
- Drainage infrastructure design

### Factors affecting infiltration
- Soil type (clay lower, sandy higher infiltration)
- Vegetation and organic matter
- Initial moisture condition
- Rainfall intensity
- Temperature
- Slope

### Measurement methods
- Single-ring infiltrometer
- Double-ring infiltrometer
- Simulation methods
- Hydrograph analysis

### Horton equation emphasis
Parameters discussed: \(f_0\), \(f_c\), \(k\), \(f(t)\), \(t\).

### Time of concentration
- Time for runoff from farthest point in catchment to reach outlet.

### Exam strategy
- Compute intervals correctly
- Convert units carefully
- Identify steady-state from tail values
- Use log transformation where linearization is required


## Meeting 3 Summary — Infiltration Capacity Curves, Linearization, and Infiltration Index

### Overview
Focused on graph interpretation, extracting model parameters, and linking cumulative infiltration to infiltration rate equations.

### Graph and best-fit interpretation
- Identify steady-state infiltration value.
- Plot transformed variables against time.
- Use straight-line best fit to obtain slope and intercept.

### Equation relation discussed
\[
F(t)=S t^{1/2}+Kt
\]
- Differentiate cumulative form to obtain infiltration rate relationship.

### Other model notes
- Kostiakov-type equations and similar derivations mentioned.
- Log transforms used for linearizing non-linear equations.

### Infiltration index concept
- Constant rate selected so rainfall excess matches measured runoff.
- Intensities above capacity/index contribute to direct runoff.


## Meeting 4 Summary — Phi Index, Runoff Depth, and Runoff Coefficient

### Overview
Covered worked examples for phi index estimation, infiltration/runoff separation, runoff coefficient, and unit-consistent calculations.

### Phi index procedure
- Assume constant infiltration threshold.
- Exclude rainfall intensities below \(\phi\).
- Recalculate until runoff condition is matched.

### Direct runoff concept
- Rainfall below \(\phi\) treated as infiltration/loss.
- Rainfall above \(\phi\) contributes to runoff excess.

### Runoff coefficient
\[
C = rac{V_{runoff}}{V_{rainfall}}
\]

### Additional exam applications
- Converting short-duration rainfall into daily averages.
- Using catchment area with rainfall depth for volume calculations.
- Distinguishing depth, volume, and discharge in answers.


## Meeting 5 Summary — Exam Review and Hydrology Answer Technique

### Overview
Exam-focused review on precipitation measurement, drizzle, condensation-to-rain process, hydrologic cycle relevance, missing-station rainfall estimation, and phi-index interpretation.

### Key points
- Precipitation measured by depth; volume from area × depth.
- Very light rainfall classified as drizzle.
- Rain forms after condensation nuclei growth and droplet coalescence.
- Missing-station rainfall can be estimated from surrounding gauges.
- Distinguish infiltration (into soil) and transpiration (release from plants).
- Use concise keyword-driven definitions in exam answers.
