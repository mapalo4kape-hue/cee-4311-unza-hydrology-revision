# Meeting 1 — Precipitation, Rain Gauges, Missing Data, and Areal Rainfall

## Overview
Revision on precipitation concepts, rainfall measurement instruments, methods for estimating missing rainfall data, and rainfall consistency checks. Strong exam focus on identifying keywords and choosing the correct method from question wording.

### Precipitation Fundamentals
- Precipitation: water falling from atmosphere to earth when gravity overcomes atmospheric resistance.
- Forms reviewed: rain, drizzle, snow, hail.
- Size references:
  - Drizzle drop diameter: **< 0.5 mm**
  - Hailstone size: about **0.5 to 5 cm**
- Why it matters: irrigation, dams, flood prediction, drainage, safety, and planning.

### Types of Precipitation
- **Convective:** heated near-surface air rises, cools, condenses, rains.
- **Orographic:** moist air forced over mountains/elevated terrain, cools and rains.
- **Cyclonic:** contrasting air masses interact; lighter air rises, condensation and rainfall follow.

### Rainfall Measurement Instruments
- Standard non-recording gauge: **ordinary rain gauge**.
- Recording gauges reviewed:
  - Tipping bucket
  - Float-type
  - Weighing-type
  - Self-recording gauges
  - Laser-based gauge (mentioned)
- Exam tip: explain **operation steps**, not just names.

### Missing Rainfall Estimation
1. **Normal ratio method** (for differing station normals)
2. **Arithmetic mean method**
3. **Inverse distance method** (closer stations weighted more)

### Consistency Check
- **Double mass curve** to detect inconsistent station records versus surrounding-station averages.
- Use slope-based correction factor if inconsistency is identified.

### Mean Areal Precipitation
- Main method emphasized: **Thiessen polygon**
  1) Plot gauges
  2) Connect neighboring gauges
  3) Draw perpendicular bisectors
  4) Form polygons
  5) Compute weighted mean by polygon area
- **Isohyetal method** also noted (equal-rainfall contour approach).

# Meeting 2 — Infiltration Fundamentals, Factors, and Horton Equation

## Overview
Tutoring review on infiltration definitions, controlling factors, measurement methods, Horton equation parameters, and exam handling of table/graph data.

### Core Definitions
- **Infiltration:** water entering soil.
- **Percolation:** downward movement after infiltration.
- **Infiltration capacity:** maximum possible infiltration rate **under given conditions**.
- **Infiltration rate:** actual rate at a given time.
- **Cumulative infiltration:** total infiltrated amount over time.
- **Steady-state infiltration:** near-constant long-time rate.

### Why It Matters
- Flood prediction/mapping
- Groundwater recharge
- Irrigation/agriculture planning
- Erosion control
- Drainage design

### Factors Affecting Infiltration
- Soil type (clay ↓, sand ↑ infiltration)
- Vegetation/organic matter (roots and pores increase infiltration)
- Initial moisture (dry soil tends to higher initial infiltration)
- Rainfall intensity
- Temperature (viscosity effects)
- Slope (steeper slope reduces opportunity time)

### Measurement Methods
- Single-ring infiltrometer
- Double-ring infiltrometer
- Simulation methods
- Hydrograph analysis

### Horton Equation Emphasis
- Parameters discussed: \(f_0\), \(f_c\), \(k\), \(f(t)\), \(t\)
- Time of concentration: travel time from hydraulically farthest point to outlet.
- Exam workflow from tables:
  - compute intervals
  - ensure unit consistency
  - identify tail/steady value
  - linearize when required (log form)

### Corrections Highlighted
- Include phrase **“under a given condition”** in infiltration-capacity definition.
- Explain vegetation impact through soil structure/channels, not photosynthesis/transpiration confusion.


# Meeting 3 — Infiltration Capacity Curves, Linearization, and Infiltration Index

## Overview
Focused on extracting infiltration parameters from plots and equations, linking cumulative infiltration to infiltration rate, and interpreting runoff-generation thresholds.

### Graph-Based Parameter Estimation
- Identify steady-state infiltration from late-time behavior.
- Plot transformed variables against time to obtain near-linear trend.
- Use best-fit straight line to obtain:
  - **slope (gradient)**
  - **intercept**

### Equation Relationships
- Cumulative infiltration form discussed:
  \[F(t)=S t^{1/2}+Kt\]
- Infiltration rate/capacity obtained by differentiation of cumulative form.
- Kostiakov-type and related forms discussed with similar conversion logic.
- Log transforms used to linearize nonlinear infiltration equations.

### Infiltration Index Concept
- Constant abstraction/loss rate selected so rainfall excess equals observed runoff.
- If rainfall intensity exceeds infiltration capacity/index, excess contributes to direct runoff.

### Practical Guidance
- Practice with same formulas in quiz-like variations.
- Memorize forms but prioritize understanding parameter meaning and graph interpretation.


# Meeting 4 — Phi Index, Runoff Depth, Runoff Coefficient, and Applied Questions

## Overview
Worked examples on phi index and runoff computations, including excluding low intensities, handling units, and solving exam-style rainfall-runoff problems stepwise.

### Phi Index Procedure
- Start with constant-loss assumption to estimate \(\phi\).
- Exclude rainfall intensities below \(\phi\) when computing direct runoff-producing rainfall.
- Iteratively adjust \(\phi\) to satisfy runoff condition.

### Direct Runoff Framing
- Conceptual relation discussed in terms of rainfall intensity adjusted by \(\phi\) over time segments.
- Key rule: intensity below \(\phi\) does not contribute to direct runoff excess.
