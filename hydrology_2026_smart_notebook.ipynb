# Meeting 1 — Precipitation, Rainfall Measurement, and Areal Rainfall

### Overview
Revision focused on precipitation fundamentals, rainfall measurement instruments, missing-data estimation, and data consistency checks.

### Core ideas
- **Precipitation**: Water falling from atmosphere to Earth when gravity overcomes resistance.
- Forms: rain, drizzle, snow, hail.
- Sizes noted:
  - Drizzle: diameter **< 0.5 mm**
  - Hail: about **0.5–5 cm**

### Why it matters in hydrology
- Flood prediction
- Irrigation planning
- Dam and drainage design
- Safety and water-resource management

### Types of precipitation
- **Convective**: heated air rises, cools, condenses, rains.
- **Orographic**: moist air forced upslope over terrain, cools, condenses.
- **Cyclonic**: contrasting air masses interact, lighter air rises and rains.

### Rain gauges discussed
- **Non-recording**: ordinary rain gauge
- **Recording**:
  - tipping bucket
  - float type
  - weighing type
  - self-recording/automatic methods
  - laser-based method (mentioned)

### Missing rainfall estimation methods
1. **Normal ratio method**

3. **Inverse distance method**

### Consistency check
- **Double mass curve**: compares cumulative target-station rainfall against cumulative average of surrounding stations.
- Break in slope indicates inconsistency/change in record behavior.

### Mean areal precipitation
- **Thiessen polygon method** (main emphasis): weighted average by influence area.
- **Isohyetal method**: contour-based equal-rainfall approach.


# Meeting 2 — Infiltration Fundamentals and Horton Equation

### Definitions
- **Infiltration**: movement of water into soil.
- **Percolation**: downward movement after infiltration.
- **Infiltration capacity**: maximum infiltration rate **under given conditions**.
- **Infiltration rate**: actual rate at a specific time.
- **Cumulative infiltration**: total infiltrated water over time.
- **Steady-state infiltration**: near-constant late-time infiltration.

### Importance
- Groundwater recharge
- Flood mapping and prediction
- Irrigation planning
- Erosion control
- Drainage design

### Factors affecting infiltration
- Soil texture/structure (sand higher; clay lower)
- Vegetation and organic matter (channels + pore development)
- Initial moisture condition (dry soil higher initial intake)
- Rainfall intensity
- Temperature (viscosity influence)
- Slope (steeper slope often lowers opportunity time)

### Measurement methods discussed
- Single-ring infiltrometer
- Double-ring infiltrometer
- Simulation-based methods
- Hydrograph-based analysis

### Horton equation focus
Parameters discussed:
- \(f_0\): initial infiltration capacity
- \(f_c\): final/steady infiltration capacity
- \(k\): decay coefficient
- \(f(t)\): infiltration at time \(t\)

### Exam strategy
- Convert units carefully (especially min ↔ hr)
- Use table intervals correctly
- Identify tail end for \(f_c\)
- Use log transformation where linear fitting is required

# Meeting 4 — Phi Index, Direct Runoff, and Runoff Coefficient

### Phi index handling
- Start with assumed constant \(\phi\).
- Exclude rainfall intensities below \(\phi\).
- Recompute until runoff condition is satisfied.

### Direct runoff idea
- Rainfall above threshold contributes to excess/runoff.
- Rainfall below threshold is treated as losses/infiltration for that period.

### Runoff coefficient
