# Lesson 1 — Precipitation Fundamentals

## Learning outcomes
By the end of this lesson, the learner should be able to:
1. Define precipitation in hydrologic terms.
2. Distinguish major precipitation forms and mechanisms.
3. Explain engineering relevance of precipitation data.

## Core content
- **Precipitation**: water released from the atmosphere and reaching the ground.
- Forms: rain, drizzle, snow, hail.
- Common size references:
  - Drizzle: droplet diameter typically < 0.5 mm
  - Hail: approximately 0.5–5 cm

## Types of precipitation
- **Convective precipitation**: uplift from surface heating.
- **Orographic precipitation**: forced uplift over terrain.
- **Cyclonic precipitation**: uplift due to air-mass interaction.

## Engineering significance
- Flood forecasting
- Irrigation planning
- Dam and drainage design
- Water-resources safety assessment

# Lesson 4 — Infiltration Models and Parameter Estimation

## Learning outcomes
1. Apply model equations to estimate infiltration behavior.
2. Linearize non-linear forms for parameter extraction.
3. Interpret slope and intercept physically.

## Key relation discussed
\[
F(t)=S\,t^{1/2}+Kt
\]
- \(F(t)\): cumulative infiltration
- Differentiate to obtain infiltration-rate relationship.

## Horton-model concepts
- Parameters commonly used: \(f_0\), \(f_c\), \(k\), \(f(t)\), \(t\)
- \(f_c\) is estimated from late-time (near-steady) behavior.

## Linearization workflow
1. Transform variables as required.
2. Plot transformed response vs time.
3. Fit best straight line.
4. Extract constants from slope/intercept.


# Lesson 5 — Phi Index, Direct Runoff, and Runoff Coefficient

## Learning outcomes
1. Solve phi-index problems stepwise.
2. Separate rainfall losses from runoff-producing excess.
3. Compute runoff coefficient using consistent units.

## Phi-index method
- Assume constant infiltration-loss rate \(\phi\).
- Exclude rainfall intensities below \(\phi\) from direct runoff excess.
- Reconcile total excess with given runoff depth.

## Runoff coefficient
\[
C = rac{V_{runoff}}{V_{rainfall}}
\]

## Unit discipline
- Convert depth, area, and time before substitution.
- Distinguish clearly among depth, discharge, and volume.
